# This workbook converts the retweet file into a usable format

```python
import pandas as pd
```

```python
df = pd.read_csv(filepath_or_buffer='../data/retweetby.csv', encoding='latin')
df.head()
```

```python
import ast
df.user = df.user.apply(lambda x: ast.literal_eval(x))
userdf = pd.DataFrame(df.user.tolist())
userdf['origin_tweetid'] = df.origin_id
userdf.head()
```

### For now we will avoid using the user entities 
this contains ulrs within there descrption as well as if they have one in the URL field within there account (already included in userdf)

```python
userentities = userdf[userdf['entities'].notnull()][['entities','id']]
pd.DataFrame(userentities.entities.tolist()).head()
```
